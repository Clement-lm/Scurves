```python
'''IMPORT DATA'''
data_panda= pd.read_excel('C:\Local\S_Curves2.xlsx', header=None)
data=data_panda.to_numpy()
#print(data)
nb_rows,nb_col = data.shape
dates=pd.to_datetime(data[0][5:],format='%d/%m/%Y')
#print(dates[5:7])
#print(data[3][5:])


list_WBS=[]
for i in range(1,nb_rows):
    if data[i][2] not in list_WBS:
        list_WBS.append(data[i][2])
list_WBS.sort()
print(list_WBS)

list_projet=[]
for i in range(1,nb_rows):
    if data[i][1] not in list_projet:
        list_projet.append(data[i][1])
list_projet.sort()
print(list_projet)
```

```python
def sigmoid(x, k, x0):
    return 100.0 / (1 + np.exp(-k * (x - x0)))

def multi_plot(x=17,y=11):
    ax=plt.figure(figsize=(x, y))
    plt.title('All S curves', fontdict={'fontweight':'bold', 'fontsize': 14})
    for i in range(1,nb_rows):
        debut,fin=(5,5)
        while data[i][debut]<1 and debut<100:
            debut+=1
        while data[i][fin]<100 and fin<100:
            fin+=1
        fin+=2
        if debut!=5:
            debut-=1
        if debut!=5:
            debut-=1
        X=dates[debut-5:fin-5]
        Y=data[i][debut:fin]
        plt.plot(X,Y,'-o',markersize=1.5,label=data[i][1]+' '+data[i][2])
    
    #plt.ylim(0,100)
    plt.grid()
    plt.yticks([0,20,40,60,80,100])
    plt.gcf().autofmt_xdate()
    plt.legend()
    plt.show()

def selection_WBS(WBS):
    ax=plt.figure(figsize=(17, 11))
    plt.title(WBS+'  S curves', fontdict={'fontweight':'bold', 'fontsize': 14})
    for i in range(1,nb_rows):
        if WBS==data[i][2]:
            debut,fin=(5,5)
            while data[i][debut]<1 and debut<100:
                debut+=1
            while data[i][fin]<100 and fin<100:
                fin+=1
            fin+=2
            if debut!=5:
                debut-=1
            if debut!=5:
                debut-=1
            X=dates[debut-5:fin-5]
            Y=data[i][debut:fin]
            plt.plot(X,Y,'-o',markersize=1.5,label=data[i][1]+' '+str(data[i][3])+' T')
            #print(i)
    
    #plt.ylim(0,100)
    plt.grid()
    plt.yticks([0,20,40,60,80,100])
    plt.yticks([])
    plt.gcf().autofmt_xdate()
    plt.legend()
    plt.show()
    
    
def selection_projet(pjt):
    ax=plt.figure(figsize=(17, 11))
    plt.title(pjt+'  S curves', fontdict={'fontweight':'bold', 'fontsize': 14})
    for i in range(1,nb_rows):
        if pjt==data[i][1]:
            debut,fin=(5,5)
            while data[i][debut]<1 and debut<100:
                debut+=1
            while data[i][fin]<100 and fin<100:
                fin+=1
            fin+=2
            if debut!=5:
                debut-=1
            if debut!=5:
                debut-=1
            X=dates[debut-5:fin-5]
            Y=data[i][debut:fin]
            plt.plot(X,Y,'-o',markersize=1.5,label=data[i][2]+' '+str(data[i][3])+' T')
            #print(i)
    
    #plt.ylim(0,100)
    plt.grid()
    plt.yticks([0,20,40,60,80,100])
    plt.gcf().autofmt_xdate()
    plt.legend()
    plt.show()    
    
def selection_projet2(pjt,option):
    ax=plt.figure(figsize=(17, 11))
    plt.title(pjt+'  S curves', fontdict={'fontweight':'bold', 'fontsize': 14})
    for i in range(1,nb_rows):
        if pjt==data[i][1]:
            debut,fin=(5,5)
            while data[i][debut]<1 and debut<100:
                debut+=1
            while data[i][fin]<100 and fin<100:
                fin+=1
            fin+=2
            if debut!=5:
                debut-=1
            if debut!=5:
                debut-=1
            Y=data[i][debut:fin]
            X=[*range(0, len(Y), 1)]
            if option==0 or option ==2:
                plt.plot(X,Y,'-o',markersize=1.5,label=str(data[i][2]))
            popt, pcov = curve_fit(sigmoid, X, Y)
            estimated_k, estimated_x0 = popt
            y_fitted = sigmoid(X, k=estimated_k, x0=estimated_x0)
            if option == 1:
                plt.plot(X+[len(Y)+1],[0]+y_fitted.tolist(),label=str(data[i][2]))
            if option == 2:
                plt.plot(X+[len(Y)+1],[0]+y_fitted.tolist(),color=plt.gca().lines[-1].get_color())
    
    #plt.ylim(0,100)
    plt.grid()
    plt.yticks([0,20,40,60,80,100])
    plt.gcf().autofmt_xdate()
    plt.legend()
    plt.show()    
       
def modele_WBS(WBS,T,option):
    ax=plt.figure(figsize=(17, 11))
    plt.title(WBS+'  S curves', fontdict={'fontweight':'bold', 'fontsize': 14})
    durees=[]
    tonnes=[]
    k_est=[]
    x0_est=[]
    for i in range(1,nb_rows):
        if WBS==data[i][2]:
            debut,fin=(5,5)
            while data[i][debut]<1 and debut<100:
                debut+=1
            while data[i][fin]<100 and fin<100:
                fin+=1
            fin+=1
            if debut!=5:
                debut-=1
                
            Y=data[i][debut:fin]
            X=[*range(0, len(Y), 1)]
            if option==0 or option ==2:
                plt.plot(X,Y,'-o',markersize=1.5,label=data[i][1]+' '+str(data[i][3])+' T')
            durees.append(len(Y))
            tonnes.append(data[i][3])
            popt, pcov = curve_fit(sigmoid, X, Y)
            estimated_k, estimated_x0 = popt
            y_fitted = sigmoid(X, k=estimated_k, x0=estimated_x0)
            x0_est.append(estimated_x0)
            k_est.append(estimated_k)
            if option == 1:
                plt.plot(X+[len(Y)+1],[0]+y_fitted.tolist(),label=data[i][1]+' '+str(data[i][3])+' T')
            if option == 2:
                plt.plot(X+[len(Y)+1],[0]+y_fitted.tolist(),color=plt.gca().lines[-1].get_color())
               
            
            '''calcul duree en fonction du tonnage'''      
               
    poly_duree=np.polyfit(tonnes,durees,1)
    duree_simulation=np.polyval(poly_duree,T)
    #print(duree_simulation)
    #print(k_est)
    #print(tonnes)
    
    poly_affine=np.polyfit(tonnes,k_est,1)
    K_SIMU=np.polyval(poly_affine,T)
    poly_affine2=np.polyfit(tonnes,x0_est,1)
    X0_SIMU=np.polyval(poly_affine2,T)   
    #K_SIMU=np.mean(k_est)
    #X0_SIMU=np.mean(x0_est)
    X=[*range(0, round(duree_simulation), 1)]
    y_fitted = sigmoid(X, k=K_SIMU, x0=X0_SIMU)
    plt.plot(X+[len(X)+1],[0]+y_fitted.tolist(),label='Simulation '+str(T)+' T',linewidth=5,color="red")
    plt.yticks([0,20,40,60,80,100])
    plt.grid()
    plt.gcf().autofmt_xdate()
    plt.legend()
    plt.show()
    
        
def modele_WBS_polynomial(WBS,T):
    ax=plt.figure(figsize=(17, 11))
    #ax.figsize=(20,15)
    plt.title(WBS+'  S curves', fontdict={'fontweight':'bold', 'fontsize': 14})
    list_poly=[]
    indice_poly=[]
    for i in range(1,nb_rows):
        if WBS==data[i][2]:
            debut,fin=(5,5)
            while data[i][debut]<1 and debut<100:
                debut+=1
            while data[i][fin]<100 and fin<100:
                fin+=1
            fin+=1
            if debut!=5:
                debut-=1
            X=dates[debut-5:fin-5]
            Y=data[i][debut:fin]
            XX=np.linspace(0,100,len(X))
            plt.plot(XX,Y,'-o',markersize=1.5,label=data[i][1]+' '+str(data[i][3])+' T')
            p1=np.polyfit(XX.tolist(),Y.tolist(),8)
            list_poly.append(p1)
            indice_poly.append(i)
    if len(list_poly)<=1:
        plt.yticks([20,40,60,80,100])
        plt.gcf().autofmt_xdate()
        plt.legend()
        plt.show()
    else :
        YY=[]
        #print(list_poly[0])
        for absc in range(101):
            poids=[]
            value=[]
            p=0
            for k in indice_poly:
                #print('k value=' +str(k))
                poids.append(data[k][3])
                value.append(np.polyval(list_poly[p],absc))
                p=+1
            poly_affine=np.polyfit(poids,value,2)
            YY.append(np.polyval(poly_affine,T))
            if YY[-1]>100:
                YY[-1]=100
            if YY[-1]<0:
                YY[-1]=0       
        #print(YY) 
        #plt.plot(list(range(101)),YY)
        popt, pcov = curve_fit(sigmoid, list(range(101)), YY)
        estimated_k, estimated_x0 = popt
        #print(estimated_k)
        y_fitted = sigmoid(list(range(101)), k=estimated_k, x0=estimated_x0)
        plt.plot(list(range(101)),y_fitted,label='Simulation '+str(T)+' T',linewidth=4)
        #plt.ylim(0,100)
        plt.grid()
        plt.yticks([0,20,40,60,80,100])
        plt.gcf().autofmt_xdate()
        plt.legend()
        plt.show()
```

```python
tonnes_widget=widgets.IntSlider(
    value=2500,
    min=500,
    max=8000,
    step=100,
    description='Tonnes :',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
wbs_widget=widgets.Dropdown(
    options=list_WBS,
    value='1.2.3',
    description='WBS :',
    disabled=False,
)

projet_widget=widgets.Dropdown(
    options=list_projet,
    value='ML',
    description='Projet :',
    disabled=False,
)

x_widget=widgets.IntSlider(
    value=17,
    min=10,
    max=25,
    step=1,
    description='Largeur :',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
y_widget=widgets.IntSlider(
    value=11,
    min=6,
    max=18,
    step=1,
    description='Hauteur :',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
option_widget=widgets.RadioButtons(
    options=[('Raw data',0), ('S-Curves Model',1), ('Both',2)],
#    value='pineapple', # Defaults to 'pineapple'
#    layout={'width': 'max-content'}, # If the items' names are long
    description='Filter :',
    disabled=False
)

#print(wbs_widget)
tab1=widgets.interactive(multi_plot,x=x_widget,y=y_widget)
tab2=widgets.interactive(selection_projet2,pjt=projet_widget,option=option_widget)
tab3=widgets.interactive(modele_WBS,WBS=wbs_widget,T=tonnes_widget,option=option_widget)


tab=widgets.Tab(children=[tab1,tab2,tab3])
tab.set_title(0,'All S curves')
tab.set_title(1,'Project selection')
tab.set_title(2,'WBS simulation')
display(tab)
```
